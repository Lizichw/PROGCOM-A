#DICCIONARIOS
Estructura de datos que almacena valores en formato llave:valor. El nombre viene de la analogía con un diccionario del mundo real, en donde la información se encuentra en un formato palabra(llave):definición(valor).

Se define de la siguiente forma:
```
d=dict()
```


O de forma alternativa:
```
d={}
```


Si se desea definir un diccionario con algunas variables:
```
d={"valor1":1}
```


Los diccionarios tienen las siguientes características:

- No poseen indexación. Para poder acceder a los valores, sólo es necesario colocar el valor entre []
- Los valores almacenados puede ser cualquier tipo de datos.
A diferencia de las listas no es mutable.


image.png



````python
#Ejemplo
#Tenemos un diccionario de superhéroes, en donde muestra su personalidad secreta y su nombre
#de superhéroe
heroes={"Nombres":["Clark Ken", "Diana Prince","Bruce Wayne"],
        "Superhéroe":["Superman","Mujer maravilla","Batman"]}
heroes["Nombres"]
````

````python
#Cómo podría imprimirse únicamente el nombre de Diana Prince?
heroes["Nombres"][1]
````

- Sólo hay una llave. Si se llegase a repetir, la segunda llave reemplazaría a la primera.

````python
heroes={"Nombres":["Clark Ken", "Diana Prince","Bruce Wayne"],
        "Superhéroe":["Superman","Mujer maravilla","Batman"],
        "Nombres":["Jhon Wilson", "Constantine"]}
heroes["Nombres"]
````

````python
#Podrías explicar este error?
heroes["nombres"]

##Explicación: Puede ser qu el erro radique en que "nombres" no esté escrito tal cual como esta en la llave ("Nombre")
````

- Las llaves pueden ser cualquier tipo de variable, lo mismo para los valores.
- Los diccionarios son ordenados a partir de Python 3.7.
- La flunción len(D) retorna el número de combinaciones llave: valor contenidas en el diccionario D

````python
len(heroes)
````
